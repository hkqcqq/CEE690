# 1. Basics of Geostatistics (25 pts)

Each observation of an environmental variable in space $z(\mathbf{x})$ can be interpreted as a single realization of a random variable $Z(\mathbf{x})$. Where $\mathbf{x}$ denotes the position in space of the observation. In a given finite region there can be an infinite number of locations $\mathbf{x}$ and thus an infinite number of random variables $Z(\mathbf{x})$. The set of all these random variables $Z(\mathbf{x})$ constitutes a random process. For the remainder of this problem assume $Z_i = Z(\mathbf{x})$ and $z_i = z(\mathbf{x})$.

A. Explain the difference between $z_i$ and $Z_i$.

B. Based on the lectures, define $Z^*$ and $Z_0$. Explain the difference between the two. 

C. Describe the meaning of $\text{Cov}(Z_i,Z_j)$, $\text{Cov}(Z_i,Z_0)$, $\text{Var}(Z_0)$.

D. The Simple Kriging system is given by $\Sigma \mathbf{w} = \mathbf{c}$. Explain each term in that equation.

E. Once $\mathbf{w}$ is computed, how can it be used to estimate $z^*$? 

F. What is second-order stationarity?

G. Assuming second-order stationarity, if we are interested in the covariance between $Z_i$ and $Z_j$, why do we concern ourselves with the semivariogram?

H. Starting from the definition of covariance, derive the relationship between covariance and semivariogram. 

A. $z_i$ is a realization or observation value at location $x$ while $Z_i$ is the distribution of all possible realizations or observation values at location $x$. And $Z_i$ is essentially a function.

B.   
$Z^* = Z^*(\mathbf{x_0}) = \sum_{i=1}^N w_i Z_i$  
$Z_0 = Z(\mathbf{x_0})$  
$Z^*$ is the distribution of estimators at location $x_0$ while $Z_0$ is the distribution of true unknown values at location $x_0$.

C.  
$\text{Cov}(Z_i,Z_j)$ means the covariance between one random variable and another random variable.  
$\text{Cov}(Z_i,Z_0)$ means the covariance between one random variable and the distribution of true unknown values at location $x_0$, which reflects how much the known random variables can inform the prediction at location $x_0$.  
$\text{Var}(Z_0)$ means the variance of the distribution of true unknown values at location $x_0$.  

D. $\mathbf{\Sigma}$ is the covariance matrix between all the observations, $\mathbf{w}$ is the vector of weights, and $\mathbf{c}$ is the vector of covariances between $Z_0$ and all observations $Z_i$.  

E. If $\mathbf{w}$ is known, we can use those weights to compute $z^*$:
$$z^* = \mathbf{w}^T\mathbf{z} = \sum_{i=1}^Nw_iz_i$$  

F. Second-order stationarity is also called weak stationarity, which means that each random variable has have a constant mean and variance.  

G. Because it is hard to determine the covariance function. Assuming second-order stationarity, it allows us to define this relationship between the semivariogram and covariance:  
$$\gamma(h) = C(0) - C(h) = \sigma^2 - \text{Cov}(Z_i,Z_j)$$
where $\gamma(h)$ is the model fit for semivariogram, $C(h)$ is the function of covariance between $Z_i$ and $Z_j$, and $h = h_{i,j} = ||\mathbf{x}_i-\mathbf{x}_j||$, $\sigma$ is the variance of the random variable $Z$. 

H. According to the defination of semivariogram:
$$\gamma_{ij} = \frac{1}{2}\text{Var}(Z_i - Z_j)$$  
Because: $$\mu_{ij} = \text{E}[Z_i - Z_j] = 0$$
Thus: $$\gamma_{ij} = \frac{1}{2}\text{Var}(Z_i - Z_j) = \frac{1}{2}\text{E}[(Z_i - Z_j - \mu_{ij})^2] = \frac{1}{2}\text{E}[(Z_i - Z_j)^2]$$
Also $R$ stands for residual, we have: $$R_i = Z_i - \mu$$
$$R_j = Z_j - \mu$$
So: $$Z_i - Z_j = Z_i - \mu - Z_j + \mu = (Z_i - \mu) - (Z_j - \mu) = R_i - R_j$$
Then: $$\gamma_{ij} = \frac{1}{2}\text{E}[(Z_i - Z_j)^2] = \frac{1}{2}\text{E}[(R_i - R_j)^2] = \frac{1}{2}\text{E}[R_i^2 - 2R_iR_j + R_j^2] = \frac{1}{2}\text{E}[R_i^2] + \frac{1}{2}\text{E}[R_j^2] - \text{E}[R_iR_j]$$
And: $$\text{E}[R_i^2] = \text{E}[R_j^2] = \text{E}[(Z_i - \mu)^2] = \text{E}[(Z_j - \mu)^2] = \text{Var}(Z_i) = \text{Var}(Z_j) = \sigma^2$$
$$\text{E}[R_iR_j] = \text{E}[(Z_i - \mu)(Z_j - \mu)] = \text{Cov}(Z_i, Z_j)$$
Thus: $$\gamma_{ij} = \frac{1}{2}\text{E}[R_i^2] + \frac{1}{2}\text{E}[R_j^2] - \text{E}[R_iR_j] = \frac{1}{2}\sigma^2 + \frac{1}{2}\sigma^2 - \text{Cov}(Z_i, Z_j) = \sigma^2 - \text{Cov}(Z_i, Z_j)$$

## A. Assemble the isotropic cloud semivariogram. Plot the results. Only consider distances up to 25000 m.

```python
import netCDF4 as nc
import numpy as np
from matplotlib import pyplot as plt
file = '/data/HW/HW4/HW4.nc'
fp = nc.Dataset(file)

sand = fp['samples']['sand'][:]
x_points = fp['samples']['x'][:]
y_points = fp['samples']['y'][:]

plt.figure(figsize=(20,20))
plt.scatter(x_points,y_points,c=sand,s=100,vmin=6,vmax=27,cmap='gnuplot2')
plt.xticks([])
plt.yticks([])
plt.title('Sand (%)',fontsize=50)
plt.show()
```

```python
#Compute euclidean distances from observatins to observations
h = ((x_points[:,np.newaxis] - x_points.T)**2 + (y_points[:,np.newaxis] - y_points.T)**2)**0.5

# Compute variogram cloud by comparing all points
gamma_c = (sand[:,np.newaxis].T - sand[:,np.newaxis])**2/2

#Reshape to 1d arrays
h = h.reshape(h.size)
gamma_c = gamma_c.reshape(gamma_c.size)

plt.figure(figsize=(15,6))

#Make the scatter plot
plt.scatter(h,gamma_c,alpha=0.01)

#Let's only consider points that are up to 25000 meters away from each other
plt.xlim([0,25000])

#Pretty up the plot
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.xlabel(r'$h_{i,j} (m)$',fontsize=40)
plt.ylabel(r'$\gamma_c(\%^2)$',fontsize=40)
plt.show()
```

## B. Having computed the cloud semivariogram, assemble the sample semivariogram using a bin size of 500 m. Plot the results. Only consider distances up to 25000 m.

```python
#Define delta
delta = 500 #m

#Define bin edges
bin_edges = np.linspace(0,25000,25000/delta+1)

#Compute the center of each bin
hd = (bin_edges[1:] + bin_edges[0:-1])/2

#Compute sample semivariogram
gamma_s = []
for i in range(hd.size):
    if i == 0:m = (h >= bin_edges[i]) & (h < bin_edges[i+1])
    else:m = (h > bin_edges[i]) & (h <= bin_edges[i+1])
    tmp = np.mean(gamma_c[m])
    gamma_s.append(tmp)
gamma_s = np.array(gamma_s)

plt.figure(figsize=(15,6))
plt.plot(hd,gamma_s,'o',markersize=20)

plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.xlabel(r'$h$',fontsize=40)
plt.ylabel(r'$\hat{\gamma}(\%^2)$',fontsize=40)
plt.show()
```

## C. Fit an exponential, spherical, and bounded linear semivariogram to the sample semivariogram. Compute the  $R^2$  of each fit then plot the results. Remember to set bounds for the curve fitting, if not it won't converge. Based on your results, which model semivariogram would you use for the region?

```python
x = fp['original_data']['x'][:]
y = fp['original_data']['y'][:]
(xs,ys) = np.meshgrid(np.arange(x.size),np.arange(y.size))
data = np.array(fp['original_data']['sand'][:])
```

### 1. Exponential semivariogram

```python
import scipy.optimize
def exponential_semivariogram(h,c,tau):
    return c * (1 - np.exp(-h/tau))

popt, pcov = scipy.optimize.curve_fit(exponential_semivariogram,hd,gamma_s,bounds=([0,1000],[100,25000]))
print('sill (c): %.2f' % popt[0])
print('range (tau): %.2f' % popt[1])
```

```python
plt.figure(figsize=(15,6))
plt.plot(hd,exponential_semivariogram(hd,*popt),lw=10)
plt.plot(hd,gamma_s,'o',markersize=20)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.title('Exponential Semivariogram',fontsize = 40)
plt.xlabel(r'$h$',fontsize=40)
plt.ylabel(r'$\gamma(\%^2)$',fontsize=40)
plt.legend([r'Sample $\gamma$',r'Modeled $\gamma$'],fontsize=25)
plt.show()
```

Ordinary kriging estimate

```python
#Define variance
var = popt[0]
#Define tau
tau = popt[1]
#Define grid positioning
x_grid = x[xs]
y_grid = y[ys]
#Let's reshape our data to a 1d array
xg1d = x_grid.reshape(x_grid.size)
yg1d = y_grid.reshape(y_grid.size)
#Compute euclidean distances from observations to points on grid
dist = ((xg1d[:,np.newaxis] - x_points.T)**2 + (yg1d[:,np.newaxis] - y_points.T)**2)**0.5
#Compute covariance matrix C
C = var*np.exp(-dist/tau).T
#Compute euclidean distances from observatins to observations
dist = ((x_points[:,np.newaxis] - x_points.T)**2 + (y_points[:,np.newaxis] - y_points.T)**2)**0.5
#Compute covariance matrix Sigma
Sigma = var*np.exp(-dist/tau).T
#Append the extra information for the langrange multipliers
#Sigma_prime
Sigma_prime = np.zeros((Sigma.shape[0]+1,Sigma.shape[1]+1))
Sigma_prime[0:-1,0:-1] = Sigma[:]
Sigma_prime[-1,0:-1] = 1
Sigma_prime[0:-1,-1] = 1
#C_prime
C_prime = np.zeros((C.shape[0]+1,C.shape[1]))
C_prime[0:-1,:] = C[:,:]
C_prime[-1,:] = 1

# Solve for the weights
W = np.linalg.solve(Sigma_prime,C_prime)

#Compute zstar
zstar = np.sum(W[0:-1,:]*sand[:,np.newaxis],axis=0)
#Compute mse 
mse = var - np.sum(W[0:-1,:]*C_prime[0:-1,:],axis=0) - W[-1,:]
ypred = zstar.reshape(x_grid.shape)
import sklearn.model_selection
print(r'R2 of exponential fit: %f' % sklearn.metrics.r2_score(data.reshape(data.size),ypred.reshape(ypred.size)))
```

```python
import matplotlib.cm as cm
vmin = 6
vmax = 27
#Plot the original and kriged data side by side
plt.figure(figsize=(20,12))
plt.subplot(122)
plt.imshow(ypred,cmap=cm.get_cmap('gnuplot2',16),vmin=vmin,vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('Estimated_Exponential',fontsize=50)
cb = plt.colorbar(shrink=0.7)
cb.ax.tick_params(labelsize=30)
plt.subplot(121)
plt.imshow(data,cmap=cm.get_cmap('gnuplot2',16),vmin=vmin,vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('Original',fontsize=50)
cb = plt.colorbar(shrink=0.7)
cb.ax.tick_params(labelsize=30)
plt.tight_layout()
plt.show()
```

### 2. Spherical semivariogram

```python
def spherical_semivariogram(h,c,a):
    return c * (3 * h / (2 * a) - 1 / 2 * (h / a)**3)**((np.trunc(h / a) - 1)**2)

popt, pcov = scipy.optimize.curve_fit(spherical_semivariogram,hd,gamma_s,bounds=([0,1000],[100,25000]))
print('sill (c): %.2f' % popt[0])
print('range (a): %.2f' % popt[1])
```

```python
plt.figure(figsize=(15,6))
plt.plot(hd,spherical_semivariogram(hd,popt[0],popt[1]),lw=10)
plt.plot(hd,gamma_s,'o',markersize=20)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.title('Spherical Semivariogram',fontsize = 40)
plt.xlabel(r'$h$',fontsize=40)
plt.ylabel(r'$\gamma(\%^2)$',fontsize=40)
plt.legend([r'Sample $\gamma$',r'Modeled $\gamma$'],fontsize=25)
plt.show()
```

## D. One of the largest limitations of Kriging is that the observation covariance matrix can become too large. Another more practical approach is to only use the closest $n$ points to each point to which we interpolate. Setting $n = 10$, perform this "localized" ordinary kriging and compare the results to those from part B. 

```python
import netCDF4 as nc
import numpy as np
from matplotlib import pyplot as plt
file = '/data/HW/HW4/HW4.nc'
fp = nc.Dataset(file)

sand = fp['samples']['sand'][:]
x_points = fp['samples']['x'][:]
y_points = fp['samples']['y'][:]

x = fp['original_data']['x'][:]
y = fp['original_data']['y'][:]
(xs,ys) = np.meshgrid(np.arange(x.size),np.arange(y.size))
data = np.array(fp['original_data']['sand'][:])

#Define variance
var = 10.0
#Define tau
tau = 1500

#Define grid positioning
x_grid = x[xs]
y_grid = y[ys]
#Let's reshape our data to a 1d array
xg1d = x_grid.reshape(x_grid.size)
yg1d = y_grid.reshape(y_grid.size)
#Compute euclidean distances from observations to points on grid
dist = ((xg1d[:,np.newaxis] - x_points.T)**2 + (yg1d[:,np.newaxis] - y_points.T)**2)**0.5

num = []
dist10 = []
for i in range(dist.shape[0]):
    min_num = []
    a = dist[i][:]
    a = a.tolist()
    min_num_index_list = np.argpartition(a, 10)[:10]
    for j in range(10):
        min_num.append(a[min_num_index_list[j]]) 
    dist10.append(min_num)
    num.append(list(min_num_index_list))
nums = np.array(num)
dist10s = np.array(dist10)
#Compute covariance matrix C
C = var*np.exp(-dist10s/tau).T

x_points10 = np.zeros((nums.shape[0], nums.shape[1]))
y_points10 = np.zeros((nums.shape[0], nums.shape[1]))
sand10 = np.zeros((nums.shape[0], nums.shape[1]))
for i in range(nums.shape[0]):
    for j in range(nums.shape[1]):
        no = nums[i][j]
        x_points10[i][j] = x_points[no]
        y_points10[i][j] = y_points[no]
        sand10[i][j] = sand[no]

W = np.zeros((11,13965))
for i in range(x_points10.shape[0]):
    xx = x_points10[i][:]
    yy = y_points10[i][:]
    #Compute euclidean distances from observatins to observations
    distsigma = ((xx[:,np.newaxis] - xx.T)**2 + (yy[:,np.newaxis] - yy.T)**2)**0.5
    #Compute covariance matrix Sigma
    Sigma = var*np.exp(-distsigma/tau).T
    #Append the extra information for the langrange multipliers
    #Sigma_prime
    Sigma_prime = np.zeros((Sigma.shape[0]+1,Sigma.shape[1]+1))
    Sigma_prime[0:-1,0:-1] = Sigma[:]
    Sigma_prime[-1,0:-1] = 1
    Sigma_prime[0:-1,-1] = 1
    #C_prime
    C_prime = np.zeros((C.shape[0]+1,1))
    C_prime[0:-1,0] = C[:,i]
    C_prime[-1,:] = 1
    # Solve for the weights
    w = np.linalg.solve(Sigma_prime,C_prime)
    W[:,i] = w[:,0]

zstars = []
mses = []
for i in range(W.shape[1]):
    zstar = 0
    msewc = 0
    for j in range(W.shape[0]-1):
        #Compute zstar
        zstar = zstar + W[j,i]*sand10[i,j]
        #Compute mse 
        msewc = msewc + W[j,i]*C[j,i]
    zstars.append(zstar)
    mse = var - msewc - W[-1,i]
    mses.append(mse)
zstars = np.array(zstars)
mses = np.array(mses)
ypred = zstars.reshape(x_grid.shape)
mse_grid = mses.reshape(x_grid.shape)

import matplotlib.cm as cm
vmin = 6
vmax = 27
#Plot the original and kriged data side by side
plt.figure(figsize=(20,12))
plt.subplot(122)
plt.imshow(ypred,cmap=cm.get_cmap('gnuplot2',16),vmin=vmin,vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('Estimated',fontsize=50)
cb = plt.colorbar(shrink=0.7)
cb.ax.tick_params(labelsize=30)
plt.subplot(121)
plt.imshow(data,cmap=cm.get_cmap('gnuplot2',16),vmin=vmin,vmax=vmax)
plt.xticks([])
plt.yticks([])
plt.title('Original',fontsize=50)
cb = plt.colorbar(shrink=0.7)
cb.ax.tick_params(labelsize=30)
plt.tight_layout()
plt.show()
```

```python
#Plot the original and kriged data side by side
plt.figure(figsize=(20,12))
plt.subplot(122)
plt.imshow(mse_grid,cmap=cm.get_cmap('viridis',16))
plt.xticks([])
plt.yticks([])
plt.title(r'$\sqrt{MSE}$ (%)',fontsize=50)
cb = plt.colorbar(shrink=0.7)
cb.ax.tick_params(labelsize=30)
plt.subplot(121)
plt.scatter(x_points,y_points,c=sand,s=100,vmin=vmin,vmax=vmax,cmap='gnuplot2')
plt.xticks([])
plt.yticks([])
plt.title('Sand (%)',fontsize=50)
cb = plt.colorbar(shrink=0.7)
cb.ax.tick_params(labelsize=30)
plt.tight_layout()
plt.show()
```

Compared the results to those from parts B and C, the estimated spatial distributions of sand content are almost identical. There is only small differnence between these two plots, which indicates that the "localized" ordinary kriging performs as well as the ordinary kriging.

## Flow Directions (also location code):

where 0 - No flow, 1 - Northwest, 2 - North, 3 - Northeast, 4 - East, 7 - West, 8 - Southwest, 9 - South, 10 - Southeast.

In this case, the sum of the code in each direction and the code in the opposite direction is a constant 11. First, I calculated the flow direction of each grid (D8 code). Then, for each specified grid, only its nearby grids whose D8 codes and their location codes to the specified grid sum up to 11 can flow into that specified grid.

```python
# Find the flow direction of each grid
d8 = np.zeros((dem.shape[0],dem.shape[1]))
for i in range(dem.shape[0]):
    for j in range(dem.shape[1]):
        demv = []
        # 4 corners
        if (i == 0) and (j == 0):
            demv.append(dem[0][1])
            demv.append(dem[1][1])
            demv.append(dem[1][0])
            demv.append(dem[0][0])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 4
            elif min_num_index_list[0] == 1: d8[i][j] = 10
            elif min_num_index_list[0] == 2: d8[i][j] = 9
            elif min_num_index_list[0] == 3: d8[i][j] = 0
        if (i == 0) and (j == 779):
            demv.append(dem[1][779])
            demv.append(dem[1][778])
            demv.append(dem[0][778])
            demv.append(dem[0][779])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 9
            elif min_num_index_list[0] == 1: d8[i][j] = 8
            elif min_num_index_list[0] == 2: d8[i][j] = 7
            elif min_num_index_list[0] == 3: d8[i][j] = 0
        if (i == 759) and (j == 0):
            demv.append(dem[758][0])
            demv.append(dem[758][1])
            demv.append(dem[759][1])
            demv.append(dem[759][0])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 2
            elif min_num_index_list[0] == 1: d8[i][j] = 3
            elif min_num_index_list[0] == 2: d8[i][j] = 4
            elif min_num_index_list[0] == 3: d8[i][j] = 0
        if (i == 759) and (j == 779):
            demv.append(dem[758][778])
            demv.append(dem[758][779])
            demv.append(dem[759][778])
            demv.append(dem[759][779])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 1
            elif min_num_index_list[0] == 1: d8[i][j] = 2
            elif min_num_index_list[0] == 2: d8[i][j] = 7
            elif min_num_index_list[0] == 3: d8[i][j] = 0
        # 4 sides
        if (i == 0) and (j > 0) and (j < 779):
            demv.append(dem[0][j + 1])
            demv.append(dem[1][j + 1])
            demv.append(dem[1][j])
            demv.append(dem[1][j - 1])
            demv.append(dem[0][j - 1])
            demv.append(dem[0][j])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 4
            elif min_num_index_list[0] == 1: d8[i][j] = 10
            elif min_num_index_list[0] == 2: d8[i][j] = 9
            elif min_num_index_list[0] == 3: d8[i][j] = 8
            elif min_num_index_list[0] == 4: d8[i][j] = 7
            elif min_num_index_list[0] == 5: d8[i][j] = 0
        if (i == 759) and (j > 0) and (j < 779):
            demv.append(dem[758][j - 1])
            demv.append(dem[758][j])
            demv.append(dem[758][j + 1])
            demv.append(dem[759][j + 1])
            demv.append(dem[759][j - 1])
            demv.append(dem[759][j])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 1
            elif min_num_index_list[0] == 1: d8[i][j] = 2
            elif min_num_index_list[0] == 2: d8[i][j] = 3
            elif min_num_index_list[0] == 3: d8[i][j] = 4
            elif min_num_index_list[0] == 4: d8[i][j] = 7
            elif min_num_index_list[0] == 5: d8[i][j] = 0
        if (j == 0) and (i > 0) and (i < 759):
            demv.append(dem[i - 1][0])
            demv.append(dem[i - 1][1])
            demv.append(dem[i][1])
            demv.append(dem[i + 1][1])
            demv.append(dem[i + 1][0])
            demv.append(dem[i][0])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 2
            elif min_num_index_list[0] == 1: d8[i][j] = 3
            elif min_num_index_list[0] == 2: d8[i][j] = 4
            elif min_num_index_list[0] == 3: d8[i][j] = 10
            elif min_num_index_list[0] == 4: d8[i][j] = 9
            elif min_num_index_list[0] == 5: d8[i][j] = 0
        if (j == 779) and (i > 0) and (i < 759):
            demv.append(dem[i - 1][778])
            demv.append(dem[i - 1][779])
            demv.append(dem[i + 1][779])
            demv.append(dem[i + 1][778])
            demv.append(dem[i][778])
            demv.append(dem[i][779])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 1
            elif min_num_index_list[0] == 1: d8[i][j] = 2
            elif min_num_index_list[0] == 2: d8[i][j] = 9
            elif min_num_index_list[0] == 3: d8[i][j] = 8
            elif min_num_index_list[0] == 4: d8[i][j] = 7
            elif min_num_index_list[0] == 5: d8[i][j] = 0
        # other internal grids
        if (i > 0) and (i < 759) and (j > 0) and (j < 779):
            demv.append(dem[i - 1][j - 1])
            demv.append(dem[i - 1][j])
            demv.append(dem[i - 1][j + 1])
            demv.append(dem[i][j + 1])
            demv.append(dem[i + 1][j + 1])
            demv.append(dem[i + 1][j])
            demv.append(dem[i + 1][j - 1])
            demv.append(dem[i][j - 1])
            demv.append(dem[i][j])
            min_num_index_list = np.argpartition(demv, 1)[:1]
            if min_num_index_list[0] == 0: d8[i][j] = 1
            elif min_num_index_list[0] == 1: d8[i][j] = 2
            elif min_num_index_list[0] == 2: d8[i][j] = 3
            elif min_num_index_list[0] == 3: d8[i][j] = 4
            elif min_num_index_list[0] == 4: d8[i][j] = 10
            elif min_num_index_list[0] == 5: d8[i][j] = 9
            elif min_num_index_list[0] == 6: d8[i][j] = 8
            elif min_num_index_list[0] == 7: d8[i][j] = 7
            elif min_num_index_list[0] == 8: d8[i][j] = 0
```
